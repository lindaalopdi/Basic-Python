## ¿Qué es POO?

Es un paradigma de la programación; un modelo básico de diseño básico de programas. En POO vamos a abstraer las cosas del mundo real y modelarlos en clases y objetos. Cada uno tiene caracteristicas que se diferencian una de otra.

## Clases y objetos


Este es un punto muy fundamental para entender la POO. En la POO todo gira en torno al objeto. Un objeto es una entidad que tiene un determinado estado, comportamiento (método) e identidad. Un objeto en POO no difiere mucho de un "objeto" del mundo real.

- El estado de un objeto (instancia) esta definido por sus atributos. No son más que datos (variables) que describen el estado actual del objeto.

- El comportamiento está especificado por los métodos. Estos métodos permiten al objeto hacer algo (andar, escribir, encenderse, etc).

- La identidad es una propiedad de la instancia que permite diferenciarla de cualquier otra. Un objeto siempre se puede diferenciar, incluso con respecto a objetos de la misma clase.

Las clases no son más que contenedores donde vamos a almacenar objetos o tambien dentro de sus propiedades es que las podemos usar como "modelo" para crear otros objetos u otras clases. En otras palabras, una clase es una entidad que define una serie de elementos que determinan un estado (datos) y un comportamiento (operaciones sobre los datos que modifican su estado).

## Constructores 

Como vimos anteriormente los métodos en una clase comientan con un guión bajo doble. Esta función tiene una relacion especial; en el caso particular de init se cumple este significado, ya que vamos a instanciar un nuevo objeto dentro de la clase. A este tipo de funciones en POO con las que inicializamos todo tipo de variable les conocemos como constructores.

```python
class Persona:
    pass #El pass ayuda a indicar que por el momento esta vacio pero puedo tener mas atributos
    def __init__(self, nombre, año):
        self.nombre = nombre 
        self.año = año 
    def descripcion(self): 
        return '{} tiene {} años'.format(self.nombre, self.año) #Usare return porque necesito que regrese valores
    def comentario(self, frase):
        return '{} dice: {}'.format(self.nombre, frase) 
    

fisico = Persona('Jose', 24)
print(fisico.nombre) #Vamos a traer un atributo

print(fisico.descripcion())#Traigo el método 
print(fisico.comentario('Hola, ¿cómo estas?'))
```

### Herencia Multiple

La herencia múltiple se refiere a crear una clase a partir de múltiples clases superiores. Al igual que en otros lenguajes, en Python como C++ y Java una clase puede derivarse más de una vez.

```python
class Base1:
    pass
class Base2:
    pass
class Derivada_Mult (Base1, Base2):
    pass

print(issubclass(Base1, Derivada_Mult))
print(issubclass(Derivada_Mult, Base2))
```

Así mismo existe algo llamado herencia multinivel, donde las características de la clase base y la clase derivada se heredan en una nueva clase derivada. Suena algo raro, pero quedara más claro con el siguiente ejemplo:

```python
class Base:
    pass
class Derivado1(Base):
    pass
class Derivado2(Derivado1):
    pass

print(issubclass(Derivado1, Base))
```

## Métodos Clase y Estáticos 

Hasta ahora hemos visto que los métodos reciben como parámetro la palabra reservada self, pero no es la unica; hay otros métodos, asi que para entender esto de las palabras reservadas (y porqué se envian como parámetro) se explicara primero las diferencias entre una clase y una instancia.

Para fines prácticos vean a una clase como una plantilla para crear objetos, algo asi como un plano o un molde. Mientras que una instancia la podemos ver de la siguiente forma; un objeto se crea usando el constructor de una clase (generalmente usamos init), una vez que es creado se le conoce como instancia de clase, sin rebuscar tanto el concepto nos referimos entonces a un objeto que se crea usando el constructor de una clase.

### Tipos de Métodos 

- Método de Clase: Comparte caracteristica con el metodo estatico que es que puede ser llamado sin crear una instancia de clase, es decir n es necesario usar init. La diferencia cae tambien en la capacidad de acceder a otros métodos y atributos de la clase sin embargo este tipo de método no tiene acceso a atributos de instancia ojo eso tenerlo en cuenta ya que ninguna instancia fue creada para poder utilizarlo.

```python
class Camarones: 
    def __init__ (self, ingredientes):
        self.ingredientes = ingredientes

#Necesitamos trabajar con listas y hay un método especial que me sirve para eso: repr 
    def __repr__ (self):
        return f'({self.ingredientes !r})' #f string nos va a ser util pues vamos a depositar varios valores

    @classmethod 
    #Elementos independientes al exterior.
    def Caldo(cls):
        return cls(['camarones', 'verduras', 'agua']) 

    @classmethod
    def Diabla(cls):
        return cls(['guajillo', 'ajo', 'jitomate', 'camarones'])

print(Camarones.Caldo())
```

- Metodo Estatico: Para poder crear un método estático es necesario anteponer @staticmethod para así indicarle a Python que el método debe de ser estático. Las características principales de un método estático es que pueden ser llamados sin tener una instancia de la clase, es decir n es necesario usar init. Además este tipo de métodos no tienen acceso al exterior, por lo cual no pueden acceder a ningún otro atributo o llamar a ninguna otra función dentro de la clase.

Este tema es un poquito complicado de entender más que todo porque no es muy común ver clases estáticas porque principalmente cuando vamos a aplicar este tipo de desarrollo son en situaciones muy muy especiales.

```python
class Terrenos: 
    def __init__ (self, ubicacion, tamaño):
        self.ubicacion = ubicacion

#Para tener referencia el tema del estático acá vamos a crear otro atributo 
        self.tamaño = tamaño
    def __repr__(self):
        return (f'Terreno en ({self.ingredientes}, 'f'{self.tamaño})')

    def area(self):
        return self.tamaño_area(self.tamaño)

#Vamos con el metodo
    @staticmethod
    def tamaño_area(A): #Refencia A de area
        return A ** 2 #terreno convenientemente cuadrado 
terreno1 = Terrenos(["Cuernavaca", "Morelos", "México"], 40) #el 40 solo sirve cuando llamo a tamaño
print(terreno1.ubicacion)
print(terreno1.tamaño_area(12), "metros cuadrados") #este valor es otro, para el metodo no para el entorno.
```

## Polimorfismo

Nos referiremos a la sobrecarga de métodos (que una subclase cuente con métodos con el mismo nombre). Quiere decir que la clase superior defina comportamientos diferentes a todo lo anterior.

Viéndolo desde el punto de vista más técnico polimorfismo en python es la capacidad que tienen los objetos en diferentes clases para usar diferentes comportamientos o atributos del mismo nombre pero con diferente valor.

```python
class Auto: 
    #voy a añadir un objeto dentro de esta clase
    ruedas = 4
    def desplazamiento(self):
        print("El auto se esta desplazando sobre 4 ruedas")
class Moto:
    ruedas = 2
    def desplazamiento(self):
        print("La moto se esta desplazando sobre 2 ruedas")
        
```

Yo tengo acá dos clases; clase auto clase moto y también tengo dos objetos: ruedas con valor de 4 y ruedas con el valor de 2. Entonces vemos que puedo tener el mismo objeto pero con diferentes valores (por ahí va el tema del polimorfismo), ahora, estas mismas clases que tienen diferentes nombres también tienen el mismo nombre en su método. Estas dos clases tienen en comun que podemos considerar que ambos son vehículos y podemos vincularlos entre sí pero hay un pequeño detalle, y es que no pueden desplazarse de la misma manera. A eso precisamente estamos llamando polimorfismo pueden ser idénticos pero en sus acciones y sus funciones son diferentes.

## Encapsulación

Los atributos de un objeto pueden ocultarse pero el modo que se van a ocultar es de modo muy superficial, principalmente para que no sean accedidos desde fuera de la definición de una clase.

Resumiendo, la encapsulación es el ocultamiento de datos del estado interno principalmente para proteger la integridad del objeto.

En si tocaremos leves métodos de seguridad para para acceder a diferentes clases.

```python
class Clientes:
    def __init__ (self):
        self.__pin = 20781 #Aqui encapsulamos una variable ok una variable que nos permita almacenar un valor 
        # en el cual no vamos a poder acceder. 
# Para desencapsularlo lo yo necesito traer una función específica
persona = Cliente()
print(persona.__pin)
```

Sale error, en cambio si lo escribo de forma normal (sin el "__") corre normal. Ahora para desencapsular necesito traer otro método

```python
class Clientes:
    def __init__ (self):
        self.__pin = 20781 
    def getpin(self):
        return self.__pin
persona = Clientes ()    
print(persona._Clientes__pin)
```
