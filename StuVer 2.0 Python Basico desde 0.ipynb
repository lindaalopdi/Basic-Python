## Fecha y tiempos

```python
import datetime as dt
import time as tm
```

```python
# Tiempo en segundos desde 1 enero 1970
tm.time()
```

```python
# Convertimos lo anterior a una fecha
dtnow = dt.datetime.fromtimestamp(tm.time())
dtnow
```

```python
# Atributos
dtnow.year, dtnow.month, dtnow.day, dtnow.hour, dtnow.minute, dtnow.second # año mes día, etc como tupla
```

```python
# Diferencia entre fechas
delta = dt.timedelta(days = 100) # crea una diferencia de 100 días
delta
```

```python
# Fecha de hoy
today = dt.date.today()
today
```

```python
# La fecha hace 100 dias
today - delta 
```

```python
# Comparar fechas
today > today-delta 
```

## Lambda y "List Comprehensions"

Lambda o una función anónima, como su nombre indica es una función sin nombre. ¿Es posible ejecutar una función sin referenciar un nombre? Pues sí, en Python podemos ejecutar una función sin definirla con def. De hecho son similares pero con una diferencia fundamental:
El contenido de una función lambda debe ser una única expresión en lugar de un bloque de acciones.

```python
# lambda: funciones rápidas
my_function = lambda a, b, c : a + b
```

```python
my_function(1, 2, 3)
```

```python
# Vamos a iterar de 0 a 999 y regresar números pares
my_list = []
for number in range(0, 1000):
    if number % 2 == 0:
        my_list.append(number)
my_list
```

```python
# lo mismo con list comprehension
my_list = [number for number in range(0,1000) if number % 2 == 0]
my_list
```
